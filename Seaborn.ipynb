# <a id = SeabornIntroduction> Seaborn Introduction </a>

Seaborn provides a high-level interface to Matplotlib, a powerful Python visualization library.

Michael Waskom says
“If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.”

## Seaborn Advantages
- Using default themes that are aesthetically pleasing.

- Setting custom color palettes.

- Making attractive statistical plots.

- Easily and flexibly displaying distributions.

- Visualizing information from matrices and DataFrames.


Seaborn is the tool for the efficient Data Exploratory Analysis

Seaborn is a complement, not a substitute, for Matplotlib.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

### <a id = lmplot> lmplot </a>

 ( same as scatter plot )
Simple syntax

<code>sns.lmplot(x="total_bill", y="tip", data=tips)</code>

where x, y and data are mandatory.

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python

g =sns.lmplot(x="total_bill", y="tip", data=tips)
```

### <a id = hue> Hue </a>

to bifurcate the data

```python
g = sns.lmplot(x="total_bill", y="tip", data=tips,hue = 'smoker')
```

## <a id = palattesdict>  Map hue colors with a dictionary </a>

palette=dict(Yes="orange", No="yellow")

this should match with the hue categories.. otherwise it will throw error

```python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette=dict(Yes="orange", No="blue"))
```


## <a id = colrow>  instead of hue in same graph, make another graph with 'col'.</a>



```python
g = sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

### bifurcate the plots with 'row' along with 'col'




```python
g = sns.lmplot(x="total_bill", y="tip",row="sex", col="time", data=tips)
```

## <a id = hueforcolor>  with 'row' and 'col' parameter, picking 'hue' for color as well.</a>


it needs merging of two columns which picks col and row, here it is 'sex' and 'time'

```python
# enumerate and double list comprehension is used. 
combined_category = [x +' '+ y for i,x in enumerate(tips.sex) for j,y in enumerate(tips.time)  if i==j ]

tips['combined_category'] = combined_category

g = sns.lmplot(x="total_bill", y="tip",row="sex", hue = 'combined_category',col="time", data=tips)

```


## <a id = heightaspect>   change height and aspect.</a>


```python
g = sns.lmplot(x="size", y="total_bill", hue="day", col="day",data=tips, height=6, aspect=.6)
```



## <a id = col_wrap>   wrapping graphs in column</a>

- col_wrap

So, if total 4 categories are there, automatically it will be wrapped in 2 cols. Which end up having 2 graphs in each row

```python
g = sns.lmplot(x="total_bill", y="tip", col="day", hue="day",data=tips, col_wrap=2, height=6)
```

## <a id = removingregline>   removing regression line</a>



```python
g = sns.lmplot(x="total_bill", y="tip", data=tips,fit_reg=False)
```

## <a id = axeslimit>  tweaking the axes limit</a>

in this plt object would be used.

```python
g = sns.lmplot(x="total_bill", y="tip", data=tips,fit_reg=False,hue='sex')
plt.xlim(-5,100)
plt.ylim(-2,20)
sns.despine(right=False,top=False)
```

### <a id = bins> bins </a>

## <a id = kde_kws> kde_kws </a> 

```python
sns.distplot(tips.total_bill,bins=50,kde_kws={"color": "k", "lw": 3, "label": "KDE"})
```

histtype

## <a id = rug_kws> rug_kws </a>

```python
sns.distplot(tips.total_bill,bins=15,rug= True,rug_kws={"color": "r", "alpha":0.3, "linewidth": 1, "height":0.2 })
```

<a id = fit_kws> fit_kws </a>

```python
from scipy import stats
sns.distplot(tips.total_bill,bins=15,fit= stats.norm,fit_kws={"color": "r", "alpha":0.3, "linewidth": 1 ,'label':'Normal'})
plt.legend()
```

```python
# color 
sns.distplot(tips.total_bill,color='red')
```

```python
j= sns.jointplot(x='total_bill',y='tip',data = tips,kind='scatter',joint_kws={'color':'red','marker' : '*' },height= 10)
j.annotate(stats.pearsonr,loc = 'best')
```

### <a id = height> Height, ratio, space </a>

- height = it is height of the joint graph
- ratio = it is ratio of marginal and joint height
- space = space between marginal and joint graph

```python
j= sns.jointplot(x='total_bill',y='tip',data = tips,height= 10,ratio=5,space = .002)
```

### <a id = markerjoint> marker, marker size, markeredgecolor, markeredgewidth, alpha </a>

# <a id = boxplot> Box Plot  </a>

- legend
- hue
-  width= width of the boxplot
-  fliersize - size of the outliers
-  linewidth - width of box border line
-  whis=1.5 -  whiskers range, for outliers
-  notch=False, - to change the shape of box

```python
plt.figure(figsize=(16,8))
sns.boxplot(x= 'day',y= 'total_bill',data = tips, hue = 'smoker')
plt.legend(loc='upper center', fontsize = 20)

```

```python
plt.figure(figsize=(16,8))
sns.boxplot(x= 'day',y= 'total_bill',data = tips, hue = 'smoker',width=.4,fliersize=7,linewidth=4,whis=4,notch=True)
plt.legend(loc='upper center', fontsize = 20)

```
