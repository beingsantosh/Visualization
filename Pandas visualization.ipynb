```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

%matplotlib inline
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df['target'] = iris.target
```

```python
df['target']=df['target'].map({0:'setosa',1:'versicolor',2:'virginica'})
```

```python
df.head()
```

### without seaborn library

```python
df['sepal length (cm)'].hist()
```

### with seaborn library

```python
import seaborn as sns
sns.set()
df['sepal length (cm)'].hist(bins=30)
```

#### another command to have hist graph

```python
df['petal length (cm)'].plot(kind='hist',bins=30)
```

## Area plot

```python
df.plot.area(alpha=.5)
```

## Bar plot

```python
plt.figure(figsize=(18,8))
df.iloc[:5,:].plot.bar(alpha=.5)
```

```python
df.iloc[:5,:].plot.bar(stacked=True)
```

### Histogram


```python
df['sepal length (cm)'].plot.hist()
```

### Line plot

```python
df.rename(columns=({'sepal length (cm)':'a','sepal width (cm)':'b','petal length (cm)':'c','petal width (cm)':'d'}),inplace=True)
```

```python
df['b']=df['b'].map(int)
df['a']=df['a'].map(int)
```

```python
df.dtypes
```

```python
df.plot.line(x='b',y='a',figsize= (17,6),lw=4)
```

### scatter plot


as it is changed to int, so only 9 points are shows in scatter.



```python
df.plot.scatter(x='a',y='b')
```


### lets run for c and d

```python
df.plot.scatter(x='c',y='d', c= 'a', figsize=(18,8), s=200, cmap= 'viridis')
```


### size vary with one variable. color also vary with one variable.    s and c parameter

```python
df.plot.scatter(x='c',y='d', c= 'a', figsize=(18,8), s=60*df['b'], cmap= 'viridis')
```

## box plot

```python
df.plot.box()
```

### hex plot

```python
df.plot.hexbin(x='c',y='d',figsize= (16,7),gridsize=25,cmap='coolwarm')
```

### KDE and density

```python
df['c'].plot.kde()
```
