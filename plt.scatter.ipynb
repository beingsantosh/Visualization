## Application of scatter plot


```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,label="marker='{0}'".format(marker))
plt.legend(loc=7)
plt.xlim(0, 1.8);
```

#### colorbar

it is to show the panel for cmap variation. 

```python
plt.figure(figsize=(15,7))
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis')
plt.colorbar();
```

## Applying scatter plot on iris dataset

```python
plt.figure(figsize=(15,7))
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=300*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## plt.plot should be preferred over plt.scatter for large datasets.

as for scatter , it works out the size and color for each point where as plot applies same properties to all the points.
