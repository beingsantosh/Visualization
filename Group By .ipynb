```python
import pandas as pd
import numpy as np
```

```python
data = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',  
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],  
        'Age':[27, 24, 22, 32,  
               33, 36, 27, 32],  
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj', 
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],  
        'Qualification':['Msc', 'MA', 'MCA', 'Phd', 
                         'B.Tech', 'B.com', 'Msc', 'MA']}  
```

```python
df = pd.DataFrame(data)
df
```

### <a id = groups> Groupby.groups</a>

```python
df.groupby(['Address','Qualification']).groups
```

### <a id = forloop> Groupby result in for loop</a>

```python
grp = df.groupby('Name') 
for name, group in grp: 
    print(name) 
    print(group) 
    print()
```

```python
grp = df.groupby(['Name', 'Qualification']) 
for name, group in grp: 
    print(name)
    print(group[['Name','Age','Qualification','Address']]) 
    
```

### <a id = getgroup> get group </a>

```python
grp = df.groupby('Name') 
grp.get_group('Jai') 
```

```python
grp = df.groupby(['Name', 'Qualification']) 
grp.get_group(('Jai', 'Msc')) 
```

### <a id = aggr> aggregate </a>

```python
grp1 = df.groupby('Name') 
  
grp1.aggregate(np.mean) 
```

```python
grp1 = df.groupby(['Name', 'Qualification']) 
  
grp1.aggregate(np.sum)
```

```python
grp = df.groupby('Name') 
  
grp['Age'].agg([np.sum, np.mean, np.std]) 
```

```python
grp = df.groupby(['Name', 'Qualification']) 
  
grp['Age'].agg([np.sum, np.mean, np.std]) 
```

### using different aggregation function by passing dictionary to aggregate

```python
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',  
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],  
        'Age':[27, 24, 22, 32,  
               33, 36, 27, 32],  
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj', 
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],  
        'Qualification':['Msc', 'MA', 'MCA', 'Phd', 
                         'B.Tech', 'B.com', 'Msc', 'MA'], 
        'Score': [23, 34, 35, 45, 47, 50, 52, 53]}
df1 = pd.DataFrame(data1)
df1
```

```python
df1.groupby('Name').agg({'Age' : ['sum','mean'], 'Score' : ['std','var']})  
  
```

```python
df1.groupby(['Name', 'Qualification']).agg({'Age' : ['sum','mean','max'], 'Score' : ['count','median','min']})  
```

### <a id = transform> Group by Transform </a>

```python
df1
```

```python
grp = df1.groupby('Name') 
sc = lambda  x: x*2
grp[['Age','Score']].transform(sc)
```

### <a id = filter> Group by filter </a>

it will put filter on the output count, for each grp.

if it has 1 rows count, it will not consider

```python
grp = df1.groupby('Name') 
grp.filter(lambda x: len(x) == 2) 
```
