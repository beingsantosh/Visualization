```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
```

# plt.hist()

```python
plt.hist(x)
plt.show()
```

## bins and alpha

> it is required to divide the entire samples in small packats as per requirement. Which helps in showcase the histogram properly.

> alpha is to recuce the transparency

```python
plt.figure(figsize=(12,6))
plt.hist(x,bins=20,alpha=.3)
plt.hist(x,bins=15,alpha = .2)
plt.show()
```

###  color and edgecolor


```python
plt.figure(figsize=(12,6))
plt.hist(x,color = 'magenta',edgecolor = 'green',bins = int(180/5))
plt.show()
```

## lable and legend

```python
plt.figure(figsize=(12,6))
plt.hist(x,color = 'magenta',edgecolor = 'green',bins = int(180/5),label = 'xyz')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.hist(x,color = 'magenta',edgecolor = 'green',bins = int(180/5),label = 'xyz',alpha=.3)
plt.hist(x,color = 'orange',edgecolor = 'black',bins = int(50/5),label = 'abc',alpha = .3)
plt.legend()
plt.show()
```

## Application - designing 4 histogram with object oriented subplots

```python
plt.figure(figsize=(18,9))
color = ['orange','red','green','blue']
for i,num in enumerate([1,5,10,15]):
    
    
    ax = plt.subplot(2,2,i+1)
    
    ax.hist(x,color = color[i],edgecolor = 'green',bins = int(100/num),label = f'Graph:{i+1}',alpha=.3  )
    
    
    ax.legend()
    ax.set_title(f'Graph with Bins bandwidth : {num}')
    ax.set_xlabel('x value',size =12)
    ax.set_ylabel('Frequency',size =12)

    
plt.tight_layout()
```

## Drawback with histogram, when it overlaps with each other, it is quite unreadable.

check below.


```python
plt.figure(figsize=(12,6))
plt.hist(x,edgecolor = 'yellow')
plt.hist(y,edgecolor = 'green')

plt.show()
```

## so to overcome this issue, sideways histogram technique is better to apply.

```python
plt.figure(figsize=(18,8))
color = ['green','orange']
labelnames = ['xyz','abc']
plt.hist([x,y],label= labelnames,color = color, edgecolor= 'black')
plt.legend(fontsize= 20)
plt.show()
```

## stacked bar

```python
plt.figure(figsize=(18,8))
color = ['green','orange']
labelnames = ['xyz','abc']
plt.hist([x,y],label= labelnames,color = color, edgecolor= 'black',stacked= True)
plt.legend(fontsize= 20)
plt.show()
```

# when density plot is needed along with histogram, then seaborn library is used.

```python
plt.figure(figsize=(18,8))
import seaborn as sns
sns.distplot(x, hist=True, kde=True, 
             bins=int(180/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```
