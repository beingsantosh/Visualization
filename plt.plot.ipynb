## <a id = 'subplothand' > 13. Subplots by hand </a>


```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],xticklabels=[],
                    ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
ax2.set_xlabel('X label')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]
                   )
ax2 = fig.add_axes([0.38, 0.6, 0.2, 0.2]
                   )

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
ax2.set_xlabel('X label')
ax1.set_title('Title')
```

## Subplot ... not subplots

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
plt.tight_layout()
```

###  <a id = 'adjustsubplot' > 14. adjustment between plots in grid </a>

hspace and wspace are the space between each plot. which is decided before actually plots are planned.

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.3, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
```

## <a id = 'sharXY' > 15. share x and y axis is important functionality</a>


otherwise whole subplot gets messy

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.25, 0.5, str((i, j)),fontsize=18)
fig
```

## <a id = 'gridspec' > 16. Gridspec .. little more complicated as compare to subplot</a>


```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

## application of gridspec

<ul>
    <li> 'invert_yaxis', invert_xaxis = completely change the graph up side down.
    <li> 'orientation' = is to change the orientation by 90 degree
    <li>    multivariate_normal
        
</ul>

```python
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(16, 10))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
